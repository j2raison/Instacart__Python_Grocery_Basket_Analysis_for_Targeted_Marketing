# Create a flag that sorts products in df_ords_prods_merge according to price.

```python
# Create price_label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Use the function
df['price_range'] = df.apply(price_label, axis=1)
```

# 2. Creation of 'busiest_days' column using if Statements with For-Loops

```python
df_ords_prods_merge['orders_day_of_week'].value_counts()
```

```python
result2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0: # Saturday
    result2.append("Busiest day 1")
  elif value == 1: # Sunday
    result2.append("Busiest day 2")
  elif value == 3: # Tuesday
    result2.append("Slower day 2")
  elif value == 4: # Wednesday
    result2.append("Slower day 1")
  else: # Monday; Thursday; Friday
    result2.append("Regularly busy")
```

```python
# Creating the 'busiest_days' column
df_ords_prods_merge['busiest_days'] = result2
```

# 4. Create a column identifying the busiest hours of the day.

```python
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

Split days in 3 periods labeled “Most orders” (busiest 2 hours: 10 and 11) “Fewest orders” (Slowest 2 hours: 3 and 4) and “Average orders” (rest of the hours).

```python
result3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value == 10:
        result3.append("Most orders")
    elif value == 11:
        result3.append("Most orders")
    elif value == 3:
        result3.append("Fewest orders")
    elif value == 3:
        result3.append("Fewest orders")
    else:
        result3.append("Average orders")
```

```python
# Adding the result list as a new column to the DataFrame
df_ords_prods_merge["busiest_period_of_day"] = result3
```
