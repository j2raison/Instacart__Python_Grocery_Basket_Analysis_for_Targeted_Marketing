# 4. Create a loyalty flag for existing customers using the transform() and loc() functions

```python
# create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Run the value_counts() function to check the values in the 'loyalty_flag' column.
df_ords_prods_merge['loyalty_flag'].value_counts()
```

```python
# check that everything was interpreted correctly and that the right flags were assigned
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(15)
```

# 5. Determining the difference between the spending habits of the three types of customers identified

```python
# Calculate the mean of the “prices” column grouped by the “loyalty_flag” column
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

# 6. Create a spending flag for each user based on the average price across all their orders

```python
# use transform() function to create a new column containing the mean amount of the “prices” column
df_ords_prods_merge['mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# create a flag that assigns a “spending” label to a user ID based on its corresponding mean spending value
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Run the value_counts() function to check the values in the 'loyalty_flag' column.
df_ords_prods_merge['spending_flag'].value_counts()
```

```python
# check that everything was interpreted correctly and that the right flags were assigned
df_ords_prods_merge[['user_id', 'spending_flag', 'order_number', 'mean_price']].head(25)
```

# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
# use transform() function to create a new column containing the median of 'days_since_prior_order' column
df_ords_prods_merge['median_days_spo'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# create a flag that assigns a “regular” label to a user ID based on its corresponding median period since prior order
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_spo'] > 20, 'regular_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_spo'] <= 20) & (df_ords_prods_merge['median_days_spo'] > 10), 'regular_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_spo'] <= 10, 'regular_flag'] = 'Frequent customer'
```

```python
# Run the value_counts() function to check the values in the 'regular_flag' column.
df_ords_prods_merge['regular_flag'].value_counts()
```

```python
# check that everything was interpreted correctly and that the right flags were assigned
df_ords_prods_merge[['user_id', 'regular_flag', 'order_number', 'median_days_spo']].head(15)
```
